```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df = pd.read_csv('Diabetes Dataset.csv')
```

```python
df
```

```python
#correlration matrix
plt.figure(figsize=(5, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm' , fmt="1.1f")
plt.title('Correlation Matrix')
```

```python
# scatter plot
plt.figure(figsize=(5, 5))
sns.scatterplot( data=df , x='BloodPressure', y='BMI' )
plt.title('Scatter Plot')
plt.xlabel('BloodPressure')
plt.ylabel('BMI')
```

```python
# Distribution plot 
fd = ['Pregnancies', 'SkinThickness', 'Age']
for i in fd:
    plt.figure(figsize=(5, 5))
    sns.histplot(df[i], kde=True, bins=20, color='blue')
    plt.ylabel("Frequency")
    plt.title(f'Distribution Plot: {i}')
```

```python
# Pie chart
outcome_ct = df['Outcome'].value_counts()
plt.figure(figsize=(5, 5))
outcome_ct.plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'red'], labels=['No Diabetes', 'Diabetes'])
plt.title('Pie Chart')
plt.ylabel('')
```

```python
# boxplot
fbox = ['SkinThickness', 'DiabetesPedigreeFunction', 'BloodPressure']
for i in fbox:
    plt.figure(figsize=(5, 5))
    sns.boxplot(x=df[i])
    plt.title(f'Boxplot: {i}')
```

```python
# mode_mean_median
print("Central Tendency:")
for i in df.columns[:9]: 
    print(f"Feature: {i}")
    print(f"  Mean: {df[i].mean()}")
    print(f"  Median: {df[i].median()}")
    print(f"  Mode: {df[i].mode()[0]}")
```

```python
# Standard and min_max
print("Dispersion:")
for col in df.columns[:9]: 
    print(f"Feature: {col}")
    print(f"  Standard Deviation: {df[col].std()}")
    print(f"  Variance: {df[col].var()}")
    print(f"  Min: {df[col].min()}")
    print(f"  Max: {df[col].max()}")
```
